```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('income.csv')
```

```python
df
```

```python
plt.scatter(df['Name'],df['Age'])
plt.,
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
```

```python
km
```

```python
y_pred =km.fit_predict(df[['Age', 'Income($)']])
```

```python
df['cluster'] = y_pred
```

```python
df
```

```python
df1 = df[df.cluster ==0]
df2 = df[df.cluster ==1]
df3 = df[df.cluster ==2]
plt.scatter(df1.Age,df1['Income($)'],color='green' )
plt.scatter(df2.Age,df2['Income($)'],color='red' )
plt.scatter(df3.Age,df3['Income($)'],color='blue' )
plt.xlabel('Age')
plt.ylabel('income')
plt.legend()
```

```python
minmax = MinMaxScaler()
```

```python
minmax.fit(df[['Income($)']])
```

```python
df['Income($)'] = minmax.transform(df[['Income($)']])
```

```python
df
```

```python
minmax.fit(df[['Age']])
```

```python
df.Age = minmax.transform(df[['Age']])
```

```python
df
```

```python
km = KMeans(n_clusters=3)
```

```python
y_pred = km.fit_predict(df[['Age','Income($)']])
```

```python
df['cluster'] = y_pred
```

```python
df
```

```python
df1 = df[df.cluster ==0]
df2 = df[df.cluster ==1]
df3 = df[df.cluster ==2]
plt.scatter(df1.Age,df1['Income($)'],color='green' )
plt.scatter(df2.Age,df2['Income($)'],color='red' )
plt.scatter(df3.Age,df3['Income($)'],color='blue' )
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],color='purple')

plt.xlabel('Age')
plt.ylabel('income')
plt.legend()
```

```python
km.cluster_centers_
```

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit_predict(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range,sse)
```
