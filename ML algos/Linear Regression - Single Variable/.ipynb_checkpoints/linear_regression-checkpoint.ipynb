```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('MFG10YearTerminationData.csv', usecols=['age','length_of_service'])
```

```python
df= df[:10]
df.head()
```

```python
plt.scatter(df.age,df.length_of_service, color='Red',marker='*')
plt.xlabel('Age (in num)')
plt.ylabel('Service(in years)')
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['age']],df.length_of_service) #fitting means traing the data
```

```python
reg.predict([[500]])
```
