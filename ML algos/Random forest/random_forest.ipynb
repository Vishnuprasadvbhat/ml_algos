# Random Forest 

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
digits = load_digits()
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits
```

```python
import pandas as pd
```

```python
df =pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
df
```

```python
x_train,x_test,y_train,y_test = train_test_split(df,digits.target,test_size=0.25)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rb =RandomForestClassifier(n_estimators=90, criterion='entropy')
```

```python
rb.fit(x_train,y_train)
```

```python
rb.score(x_test,y_test)
```
