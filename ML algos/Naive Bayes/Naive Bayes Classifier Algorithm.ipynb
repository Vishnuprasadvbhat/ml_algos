```python
import pandas as pd
df= pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df
```

```python
target = df.Survived
target
```

```python
input = df.drop('Survived',axis= 'columns' )
```

```python
dummies = pd.get_dummies(input.Sex)
dummies
```

```python
input= pd.concat([input,dummies],axis='columns')
```

```python
input
```

```python
input.drop(['Sex'],axis='columns',inplace= True)
```

```python
input.columns[input.isna().any()]
```

```python
input.isnull().sum()
```

```python
input.Age =  input.Age.fillna(input.Age.mean())
```

```python
input
```

```python
input.mad
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(input,target,test_size=0.3)
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
node = GaussianNB()
```

```python
node.fit(x_train,y_train)
```

```python
node.score(x_test,y_test)
```
